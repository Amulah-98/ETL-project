```python
import psycopg2
from psycopg2.extensions import ISOLATION_LEVEL_AUTOCOMMIT

def create_database():
    # Define connection parameters
    hostname = "localhost"
    dbname = "postgres"
    username = "postgres"
    password = "password"  # Replace with your actual PostgreSQL password

    # Establish connection to the default database
    conn = psycopg2.connect(
        host=hostname,
        dbname=dbname,
        user=username,
        password=password
    )
    conn.set_isolation_level(ISOLATION_LEVEL_AUTOCOMMIT)

    # Create cursor object
    cur = conn.cursor()

    # Create a new database
    cur.execute("CREATE DATABASE accounts_database;")

    # Close cursor and connection to default database
    cur.close()
    conn.close()

    # Connect to the newly created database
    conn = psycopg2.connect(
        host=hostname,
        dbname="accounts_database",
        user=username,
        password=password
    )
    cur = conn.cursor()
    
    return cur, conn

# Call the function to create database
cur, conn = create_database()
```
